```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df1 = pd.read_csv("rain.csv",encoding="cp949")

df2 = df1.dropna()
df2arr = np.array(df2)

precipitation = []
sum_precipitation = []
for area in df2arr:
    arr = [area[0]]
    for month in range(1, 13):
        arr.append(area[month])
    precipitation.append(arr)
    sum_precipitation.append([area[0], area[13]])

    
sum_precipitation.sort(key=lambda x : x[1],reverse = True)

X = []
for i in sum_precipitation:
   X.append(i[1])

Y = []
for i in sum_precipitation:
    Y.append(i[0])
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
    
plt.rcParams['axes.unicode_minus'] = False   

data = X[0:7]
data1 = Y[0:7]
ind = np.arange(7)
plt.figure(figsize=(10, 5))

plt.bar(range(len(data)), data, color = 'green')
plt.xticks(ind, data1)
plt.show
```
